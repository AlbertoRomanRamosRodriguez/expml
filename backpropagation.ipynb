# Backpropagation implementation testing

```python
from arch_imp.nn.NeuralNetwork import NeuralNetwork
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn import datasets

import numpy as np
```

```python
def train_network(X, y, operation,shape = [2,2,1], X_test = None, y_test = None, epochs = 20000):
    

    nn = NeuralNetwork(shape, alpha=0.5)
    nn.fit(X,y,epochs=epochs)

    print(nn)
    
    if operation == 'XOR':
        X_test = X
        y_test = y

        for (x,target) in zip(X_test,y_test):
            pred = nn.predict(x)[0][0]
            step = 1 if pred > 0.5 else 0
            print(f'[INFO] data = {x}, ground-truth={target[0]}, pred ={pred}, step={step}')
    else:
        predictions = nn.predict(X_test).argmax(axis=1)
        print(classification_report(y_test.argmax(axis=1), predictions))
```

## XOR dataset

```python
operation = 'XOR'
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])
shape = [2,2,1]

train_network(X, y, operation,shape)
```

## MNIST Dataset

```python
print("[INFO] loading MNIST (sample) dataset")
digits = datasets.load_digits()
data = digits.data.astype('float')
data = (data - data.min()) / (data.max()-data.min())
print(f"[INFO] samples: {data.shape[0]}, dim: {data.shape[1]} ")

(trainX, testX, trainY, testY) = train_test_split(data, digits.target, test_size=0.25)

trainY = LabelBinarizer().fit_transform(trainY)
testY = LabelBinarizer().fit_transform(testY)

operation = 'MNIST'
shape = [trainX.shape[1], 32, 16, 10]

train_network(trainX, trainY, operation,shape,testX,testY,1000)
```
